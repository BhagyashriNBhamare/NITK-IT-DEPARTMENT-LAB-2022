```python
import csv
import random
import math
import copy
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
def readCSV(filename):
	fields =[]
	records =[]
	with open(filename,"r") as csvFile:
		csvReader = csv.reader(csvFile)

		for row in csvReader:
			records.append(row)


	fields = records[0]
	records = records[1:]
	for i in range(len(records)):
		for j in range(1,len(records[0])):
			records[i][j] = float(records[i][j])
	return records


def dist(arr1, arr2):
    temp_dist = 0
    for i in range(len(arr1)):
        temp_dist = temp_dist + ((float(arr1[i])-float(arr2[i]))**2)


    temp_dist = temp_dist**(0.5)

    return temp_dist

def scalar(k, arr1):
    temp_prod=[]
    for i in range(len(arr1)):
        temp_prod.append(float(arr1[i])*k)

    return temp_prod

def vecAdd(arr1, arr2):
    temp_sum = list()
    for i in range(len(arr1)):
        temp_sum.append(float(arr1[i])+float(arr2[i]))

    return temp_sum

def vecDiff(arr1, arr2):
    diff = 0
    for i in range(len(arr1)):
        diff = diff+arr1[i]-arr2[i]

    return diff


def getAccuracy(centre, records, no_of_records):
    a={}
    d1={}
    d1[0]="setosa"
    d1[1]="versicolor"
    d1[2]="virginica"
    c1={}
    c1[0]=[]
    c1[1]=[]
    c1[2]=[]
    acc1=0
    correct1=0
    total=0
    for x in records:
        total=total+1
        if dist(x[1:],centre[0])<=dist(x[1:],centre[1]):
            if dist(x[1:],centre[0])<=dist(x[1:],centre[2]):
                c1[0].append(x)
                if x[0]==d1[0]:
                    correct1=correct1+1
            else:
                c1[2].append(x)
                if x[0]==d1[2]:
                    correct1=correct1+1
        else:
            if dist(x[1:],centre[1])<=dist(x[1:],centre[2]):
                c1[1].append(x)
                if x[0]==d1[1]:
                    correct1=correct1+1
            else:
                c1[2].append(x)
                if x[0]==d1[2]:
                    correct1=correct1+1
    acc1=(correct1*1.0)/total
#     print(acc1,correct1,total)
    a[acc1]=[d1,c1]
    d2={}
    d2[0]="setosa"
    d2[1]="virginica"
    d2[2]="versicolor"
    c2={}
    c2[0]=[]
    c2[1]=[]
    c2[2]=[]
    acc2=0
    correct2=0
    for x in records:
        if dist(x[1:],centre[0])<=dist(x[1:],centre[1]):
            if dist(x[1:],centre[0])<=dist(x[1:],centre[2]):
                c2[0].append(x)
                if x[0]==d2[0]:
                    correct2=correct2+1
            else:
                c2[2].append(x)
                if x[0]==d2[2]:
                    correct2=correct2+1
        else:
            if dist(x[1:],centre[1])<=dist(x[1:],centre[2]):
                c2[1].append(x)
                if x[0]==d2[1]:
                    correct2=correct2+1
            else:
                c2[2].append(x)
                if x[0]==d2[2]:
                    correct2=correct2+1
    acc2=(correct2*1.0)/total 
#     print(acc2,correct2,total)
    a[acc2]=[d2,c2]
    d3={}
    d3[0]="virginica"
    d3[1]="setosa"
    d3[2]="versicolor"
    c3={}
    c3[0]=[]
    c3[1]=[]
    c3[2]=[]
    acc3=0
    correct3=0
    for x in records:
        if dist(x[1:],centre[0])<=dist(x[1:],centre[1]):
            if dist(x[1:],centre[0])<=dist(x[1:],centre[2]):
                c3[0].append(x)
                if x[0]==d3[0]:
                    correct3=correct3+1
            else:
                c3[2].append(x)
                if x[0]==d3[2]:
                    correct3=correct3+1
        else:
            if dist(x[1:],centre[1])<=dist(x[1:],centre[2]):
                c3[1].append(x)
                if x[0]==d3[1]:
                    correct3=correct3+1
            else:
                c3[2].append(x)
                if x[0]==d3[2]:
                    correct3=correct3+1
    acc3=(correct3*1.0)/total 
#     print(acc3,correct3,total)
    a[acc3]=[d3,c3]
    d4={}
    d4[0]="virginica"
    d4[1]="versicolor"
    d4[2]="setosa"
    c4={}
    c4[0]=[]
    c4[1]=[]
    c4[2]=[]
    acc4=0
    correct4=0
    for x in records:
        if dist(x[1:],centre[0])<=dist(x[1:],centre[1]):
            if dist(x[1:],centre[0])<=dist(x[1:],centre[2]):
                c4[0].append(x)
                if x[0]==d4[0]:
                    correct4=correct4+1
            else:
                c4[2].append(x)
                if x[0]==d4[2]:
                    correct4=correct4+1
        else:
            if dist(x[1:],centre[1])<=dist(x[1:],centre[2]):
                c4[1].append(x)
                if x[0]==d4[1]:
                    correct4=correct4+1
            else:
                c4[2].append(x)
                if x[0]==d4[2]:
                    correct4=correct4+1
    acc4=(correct4*1.0)/total 
#     print(acc4,correct4,total)
    a[acc4]=[d4,c4]
    d5={}
    d5[0]="versicolor"
    d5[1]="virginica"
    d5[2]="setosa"
    c5={}
    c5[0]=[]
    c5[1]=[]
    c5[2]=[]
    acc5=0
    correct5=0
    for x in records:
        if dist(x[1:],centre[0])<=dist(x[1:],centre[1]):
            if dist(x[1:],centre[0])<=dist(x[1:],centre[2]):
                c5[0].append(x)
                if x[0]==d5[0]:
                    correct5=correct5+1
            else:
                c5[2].append(x)
                if x[0]==d5[2]:
                    correct5=correct5+1
        else:
            if dist(x[1:],centre[1])<=dist(x[1:],centre[2]):
                c5[1].append(x)
                if x[0]==d5[1]:
                    correct5=correct5+1
            else:
                c5[2].append(x)
                if x[0]==d5[2]:
                    correct5=correct5+1
    acc5=(correct5*1.0)/total 
#     print(acc5,correct5,total)
    a[acc5]=[d5,c5]
    d6={}
    d6[0]="versicolor"
    d6[1]="setosa"
    d6[2]="virginica"
    c6={}
    c6[0]=[]
    c6[1]=[]
    c6[2]=[]
    acc6=0
    correct6=0
    for x in records:
        if dist(x[1:],centre[0])<=dist(x[1:],centre[1]):
            if dist(x[1:],centre[0])<=dist(x[1:],centre[2]):
                c6[0].append(x)
                if x[0]==d6[0]:
                    correct6=correct6+1
            else:
                c6[2].append(x)
                if x[0]==d6[2]:
                    correct6=correct6+1
        else:
            if dist(x[1:],centre[1])<=dist(x[1:],centre[2]):
                c6[1].append(x)
                if x[0]==d6[1]:
                    correct6=correct6+1
            else:
                c6[2].append(x)
                if x[0]==d6[2]:
                    correct6=correct6+1
    acc6=(correct6*1.0)/total 
#     print(acc6,correct6,total)
    a[acc6]=[d6,c6]
    f=0
    for x in a:
        if x>f:
            f=x
            ans=[f,a[x]]
    return ans






def draw(data,centres,c):
    a1=[]
    b1=[]
    a2=[]
    b2=[]
    a3=[]
    b3=[]
    for x in data[0]:
        a1.append(x[1])
        b1.append(x[2])
    for x in data[1]:
        a2.append(x[1])
        b2.append(x[2])
    for x in data[2]:
        a3.append(x[1])
        b3.append(x[2])
    red_patch = mpatches.Patch(color='red', label='Centroids')
    black = mpatches.Patch(color='yellow', label='cluster 1 '+str(c[0]))
    red_patc = mpatches.Patch(color='blue', label='cluster 2 '+str(c[1]))
    re = mpatches.Patch(color='pink', label='cluster 3 '+str(c[2]))
    plt.legend(handles=[black,red_patc,re,red_patch])
    plt.scatter(a1,b1,color='yellow')
    plt.scatter(a2,b2,color='blue')
    plt.scatter(a3,b3,color='pink')
    plt.scatter([centres[0][1],centres[1][1],centres[2][1]],[centres[0][2],centres[1][2],centres[2][2]],s=50,c='red',label='Centroids')
    plt.xlabel('sepel') 
    plt.ylabel('petal') 
    plt.show()

def call(records,old_centroids,i,j,m):
    s=0
    a=dist(records[i][1:],old_centroids[j])
    
    for x in range(3):
        s=s+(a/dist(records[i][1:],old_centroids[x]))**(2/(m-1))
    return s
def fuzzyCMeans(c, m, no_of_attr, no_of_iter, records, no_of_records):

    old_centroids = [[random.uniform(0,90) for j in range(no_of_attr)] for i in range(c)]
    new_centroids = [[0 for j in range(no_of_attr)] for i in range(c)]

    membership_matrix = [[0 for j in range(c)] for i in range(no_of_records)]

    for num in range(0, no_of_iter):
        for i in range(0, no_of_records):
            membership_matrix[i][0] = 1/call(records,old_centroids,i,0,2)
            membership_matrix[i][1] = 1/call(records,old_centroids,i,1,2)
            membership_matrix[i][2] = 1/call(records,old_centroids,i,2,2)
        error =0
#         for x in range(c):
#             for y in range(len()):
#                 for n in range(1:4):
#                     a=a+records[m][n]*(membership_matrix[m][x]**m)
#                     b=b+(membership_matrix[m][x]**m)
#                 p=a/b
#                 new_centroids[x][n]=p
#         old_centroids = copy.deepcopy(new_centroids)
        for j in range(0, c):
            denom = 0
            numer = [0 for numb in range(no_of_attr)]
            for i in range(0, no_of_records):
                denom = denom+(membership_matrix[i][j]**m)
                numer = vecAdd(numer, scalar((membership_matrix[i][j]**m), records[i][1:]))

            if(denom==0):
                denom =1
                break

            denom = denom**(-1)
            new_centroids[j] = scalar(denom, numer)
            for x in new_centroids[j]:
                x = math.floor(x)

            error = vecDiff(new_centroids[j], old_centroids[j])
        old_centroids = copy.deepcopy(new_centroids)


    return new_centroids, membership_matrix

def main():
    filename = "IRIS.csv"
    records = readCSV(filename)
    c = 3
    m = 2
    no_of_iter = 100
    no_of_attr = len(records[0])-1
    no_of_records = len(records)
    centres, membership_matrix = fuzzyCMeans(c, 2, no_of_attr, no_of_iter, records, no_of_records)
    accuracy = getAccuracy(centres, records, no_of_records)
    print("Result Accuracy fuzzy : ",accuracy[0]*100,"\n \n ")
    c=accuracy[1].pop(0)
    data=accuracy[1].pop(0)

    m=[[0,0,0]for x in range(3)]
    for x in data[0]:
        if x[0]==c[0]:
            m[0][0]=m[0][0]+1
        else:
            if c[1]==x[0]:
                y=1
            else:
                y=2
            m[0][y]=m[0][y]+1
    for x in data[1]:
        if x[0]==c[1]:
            m[1][1]=m[1][1]+1
        else:
            if c[0]==x[0]:
                y=0
            else:
                y=2
            m[1][y]=m[1][y]+1

    for x in data[2]:
        if x[0]==c[2]:
            m[2][2]=m[2][2]+1
        else:
            if c[1]==x[0]:
                y=1
            else:
                y=0
            m[2][y]=m[2][y]+1
    print("Graphical representation of clusters \n \n\n ")
    draw(data,centres,c)
    print("Conclusion Matrix")
    df_cm = pd.DataFrame(m, range(3), range(3))
    sn.set(font_scale=1.4)
    sn.heatmap(df_cm, annot=True, annot_kws={"size": 16}) # font size
    plt.show()
    

if __name__ == '__main__':
    main()
```
